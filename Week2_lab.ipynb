```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

#Exercise 1


```python
areas = [30, 40, 32, 28]
prices = [33000, 40000, 30000, 28000]

print(areas)
print(prices)
```

```python
areas_np = np.array(areas)
prices_np = np.array(prices)

print(areas_np)
print(np.mean(areas_np))
print(np.median(areas_np))
```

```python
print(max(prices_np))
print(min(prices_np))
```

```python
plt.scatter(areas_np, prices_np)
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.title('Relationship between Area and Price')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

X = areas_np.reshape(-1, 1)
y = prices_np

model = LinearRegression()
model.fit(X, y)

print('Slope:', model.coef_[0])
print('Intercept:', model.intercept_)
```

```python
new_area = 70
new_area_1 = 110
new_area_2 = 120
predicted_price = model.predict([[new_area]])
predicted_price_1 = model.predict([[new_area_1]])
predicted_price_2 = model.predict([[new_area_2]])
print('Predicted price for area', new_area, 'm2:',predicted_price[0])
print('Predicted price for area', new_area_1,'m2',predicted_price_1[0])
print('Predicted price for area', new_area_2,'m2',predicted_price_2[0])
```
